## Ingest Products

```python
products_df = (
    spark.read
         .option("header", "true")
         .option("inferSchema", "true")    
         .csv("abfss://uploaded-data@datastorage00578.dfs.core.windows.net/Instacart/products.csv")
)

products_df.printSchema()
```

```python
display(products_df.limit(10))
```

```python
from pyspark.sql.types import StructType, StructField, IntegerType, StringType, DoubleType

products_schema = StructType([StructField('product_id', IntegerType(), False),
                              StructField('product_name', StringType(), True),
                              StructField('aisle_id', IntegerType(), True),
                              StructField('department_id', IntegerType(), True)])


products_df = spark.read.schema(products_schema) \
                      .option("header", "true") \
                      .csv("abfss://uploaded-data@datastorage00578.dfs.core.windows.net/Instacart/products.csv")
```

```python
display(products_df.limit(5))
```

```python
# Write as Delta
products_df.write.format("delta") \
               .mode("overwrite") \
               .option("path", "abfss://processed-data@datastorage00578.dfs.core.windows.net/Instacart/bronze/products") \
               .saveAsTable("instacart.bronze.products")
```
