## Ingest Order-Products-Prior

```python
ord_pro_prior_df = (
    spark.read
         .option("header", "true")
         .option("inferSchema", "true")    
         .csv("abfss://uploaded-data@datastorage00578.dfs.core.windows.net/Instacart/order_products__prior.csv")
)

ord_pro_prior_df.printSchema()
```

```python
display(ord_pro_prior_df.limit(10))
```

```python
from pyspark.sql.types import StructType, StructField, IntegerType, StringType, DoubleType

ord_pro_prior_schema = StructType([StructField('order_id', IntegerType(), True),
                                      StructField('product_id', IntegerType(), True),
                                      StructField('add_to_cart_order', IntegerType(), True),
                                      StructField('reordered', IntegerType(), True) ])


ord_pro_prior_df = spark.read.schema(ord_pro_prior_schema) \
                      .option("header", "true") \
                      .csv("abfss://uploaded-data@datastorage00578.dfs.core.windows.net/Instacart/order_products__prior.csv")
```

```python
display(ord_pro_prior_df.limit(5))
```

```python
# Write as Delta
ord_pro_prior_df.write.format("delta") \
               .mode("overwrite") \
               .option("path", "abfss://processed-data@datastorage00578.dfs.core.windows.net/Instacart/bronze/order_products_prior") \
               .saveAsTable("instacart.bronze.order_products_prior")
```
