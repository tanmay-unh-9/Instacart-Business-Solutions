## Ingest Departments

```python
departments_df = (
    spark.read
         .option("header", "true")
         .option("inferSchema", "true")    
         .csv("abfss://uploaded-data@datastorage00578.dfs.core.windows.net/Instacart/departments.csv")
)

departments_df.printSchema()
```

```python
display(departments_df.limit(10))
```

```python
from pyspark.sql.types import StructType, StructField, IntegerType, StringType

departments_schema = StructType([StructField('department_id', IntegerType(), False),
                                 StructField('department', StringType(), True)])

departments_df = spark.read.schema(departments_schema) \
                      .option("header", "true") \
                      .csv("abfss://uploaded-data@datastorage00578.dfs.core.windows.net/Instacart/departments.csv")
```

```python
display(departments_df.limit(5))
```

```python
# Write as Delta
departments_df.write.format("delta") \
               .mode("overwrite") \
               .option("path", "abfss://processed-data@datastorage00578.dfs.core.windows.net/Instacart/bronze/departments") \
               .saveAsTable("instacart.bronze.departments")
```
