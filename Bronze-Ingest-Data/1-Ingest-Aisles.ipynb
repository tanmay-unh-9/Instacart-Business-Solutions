## Ingest Aisles

```python
aisles_df = (
    spark.read
         .option("header", "true")
         .option("inferSchema", "true")     # or define a schema explicitly for production
         .csv("abfss://uploaded-data@datastorage00578.dfs.core.windows.net/Instacart/aisles.csv")
)

aisles_df.printSchema()
```

```python
from pyspark.sql.types import StructType, StructField, IntegerType, StringType

aisles_schema = StructType([StructField('aisle_id', IntegerType(), False),
                            StructField('aisle', StringType(), True)])

aisles_df = spark.read.schema(aisles_schema) \
                      .option("header", "true") \
                      .csv("abfss://uploaded-data@datastorage00578.dfs.core.windows.net/Instacart/aisles.csv")
```

```python
display(aisles_df.limit(5))
```

```python
# Write as Delta
aisles_df.write.format("delta") \
               .mode("overwrite") \
               .option("path", "abfss://processed-data@datastorage00578.dfs.core.windows.net/Instacart/bronze/aisles") \
               .saveAsTable("instacart.bronze.aisles")
```
